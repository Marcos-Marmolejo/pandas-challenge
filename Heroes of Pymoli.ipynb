```python
import pandas as pd
```

```python
file = "lol.csv"
lol_df = pd.read_csv(file)
lol_df.head()
```

```python
#Display the total number of players
count = len(lol_df["SN"].unique())
count
```

```python
count = len(lol_df["Item ID"].unique())
count
```

```python
#Purchasing Analysis (Total)
x= len(lol_df["Item ID"].unique())
y= lol_df["Price"].mean()
w= lol_df["SN"].count()
z= lol_df["Price"].sum()
frame_df = pd.DataFrame({
    "Number of Unique Items":[x],
    "avg price":[y],
    "Number of Purchases":[w],
    "Total revenue":[z]
})
frame_df
```

```python
# Gender Demographics
gender_select = lol_df[['Gender','SN']]
data_sum = gender_select.groupby('Gender').nunique()
data = pd.DataFrame(data_sum)
x = data["SN"].sum()
data["Percentage of Players"] = (data["SN"] / x)*100
data
```

```python
# Purchasing Analysis (Gender)

kick_4 = lol_df.loc[lol_df["Gender"] == 'Female', [
    "Purchase ID", "SN","Item ID" ,"Price"]]

kick_5 = lol_df.loc[lol_df["Gender"] == 'Male', [
    "Purchase ID", "SN","Item ID" ,"Price"]]

kick_6 = lol_df.loc[lol_df["Gender"] == 'Other / Non-Disclosed', [
    "Purchase ID", "SN","Item ID" ,"Price"]]

x= len(kick_4["SN"].unique())
y= kick_4["Price"].mean()
w= kick_4["SN"].count()
z= kick_4["Price"].sum()
a = z/x

x1= len(kick_5["SN"].unique())
y1= kick_5["Price"].mean()
w1= kick_5["SN"].count()
z1= kick_5["Price"].sum()
a1 = z1/x1

x2= len(kick_6["SN"].unique())
y2= kick_6["Price"].mean()
w2= kick_6["SN"].count()
z2= kick_6["Price"].sum()
a2 = z2/x2

frame_df = pd.DataFrame({
    "Gender":["Female","Male","Other / Non-Disclosed"],
    "Purchase Count":[w,w1,w2],
    "avg price":[y,y1,y2],
    "Total revenue":[z,z1,z2],
    "Avg Total Purchase per Person":[a,a1,a2]
})
frame_df
```

```python
#Crando bins para data frame
lol_df['bin'] = pd.cut(lol_df['Age'], bins=[0,9,14,19,24,29,34,39,50],labels=['<10', '10-14', '15-19','20-24','25-29','30-34','35-39','40+'])
#Contando jugadores por bin
dbin = lol_df[['bin','SN']].groupby(['bin']).nunique()
dbin =dbin['SN']
dbin= dbin.to_frame()
dbin = dbin.reset_index(level=0)
dbin= dbin.rename(columns={"":"Bin",
                                "SN":"Count of SN"})

dbin['Percentage of Players'] = (dbin['Count of SN']/576)*100

dbin
```

```python
#Purchasing Analysis (Age)
#Obtenemos promedio,total y numero de players
df_binPurchase = lol_df[['bin','Purchase ID']].groupby(['bin']).nunique()
df_binPurchaseprice = lol_df[['bin','Price']].groupby(['bin']).mean()
df_binPurchasetotalprice = lol_df[['bin','Price']].groupby(['bin']).sum()
df_bincount = lol_df[['bin','SN']].groupby(['bin']).nunique()
avgTbin=df_binPurchasetotalprice['Price']/df_bincount['SN']



#Agregamos nuevas columnas con los datos que nos piden
df_binPurchase['Avg Price']= df_binPurchaseprice['Price']
df_binPurchase['Total Revenue']= df_binPurchasetotalprice['Price']
df_binPurchase['Avg Total Purchase per Person'] = avgTbin
df_binPurchase = df_binPurchase.rename(columns={"Purchase ID":"Purchase Count"})
df_binPurchase=df_binPurchase[['Purchase Count','Avg Price','Total Revenue','Avg Total Purchase per Person']]
df_binPurchase
```

```python
# Top Spenders

#avg price
file_select = lol_df[['SN','Price']]
grouped_usa_df = file_select.groupby(['SN']).mean()

#number purches
file = lol_df[['SN','Purchase ID']]
grouped_df = file.groupby(['SN']).nunique()
data = grouped_df[['Purchase ID']]

merge_table = pd.merge(grouped_usa_df, data, on="SN")
merge_table["Total Purchase Value"] = (merge_table["Price"] * merge_table["Purchase ID"])
dat=merge_table.sort_values(by='Total Purchase Value', ascending=False)
dat = dat.rename(columns={"Purchase ID":"Purchase Count"})
dat[:5] 
```

```python
# Most Popular Items

#avg price
file_select = lol_df[['Item Name','Price']]
grouped_usa_df = file_select.groupby(['Item Name']).mean()

#number purches
file = lol_df[['Item Name','Purchase ID']]
grouped_df = file.groupby(['Item Name']).nunique()
data = grouped_df[['Purchase ID']]
#dat=data.sort_values(by='Purchase ID', ascending=False)
#dat

merge_table = pd.merge(grouped_usa_df, data, on="Item Name")
#merge_table

merge_table["Total Purchase Value"] = (merge_table["Price"] * merge_table["Purchase ID"])
#merge_table
dat1=merge_table.sort_values(by='Purchase ID', ascending=False)
dat1 = dat1.rename(columns={"Purchase ID":"Purchase Count"})
dat1[:5] 
```

```python
# Most Profitable Items

#avg price
file_select = lol_df[['Item Name','Price']]
grouped_usa_df = file_select.groupby(['Item Name']).mean()

#number purches
file = lol_df[['Item Name','Purchase ID']]
grouped_df = file.groupby(['Item Name']).nunique()
data = grouped_df[['Purchase ID']]
#data

merge_table = pd.merge(grouped_usa_df, data, on="Item Name")
#merge_table

merge_table["Total Purchase Value"] = (merge_table["Price"] * merge_table["Purchase ID"])

dat=merge_table.sort_values(by='Total Purchase Value', ascending=False)
dat = dat.rename(columns={"Purchase ID":"Purchase Count"})
dat[:5] 
```
